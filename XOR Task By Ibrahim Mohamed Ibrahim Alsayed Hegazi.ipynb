```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def logic_gate(w1, w2, b, x1, x2):
    return sigmoid(w1 * x1 + w2 * x2 + b)

def xor_gate(x1, x2):
    # First layer: OR and NAND
    or_output = logic_gate(20, 20, -10, x1, x2)
    nand_output = logic_gate(-20, -20, 30, x1, x2)
    # Second layer: AND (combines OR and NAND)
    xor_output = logic_gate(20, 20, -30, or_output, nand_output)
    return np.round(xor_output)

# Test XOR gate
print("XOR Gate Truth Table:")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"{x1}, {x2}: {xor_gate(x1, x2)}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# XOR dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y = np.array([[0], [1], [1], [0]], dtype=np.float32)  # XOR outputs

# Build the model
model = Sequential([
    Dense(2, input_dim=2, activation='relu'),  # Hidden layer (2 neurons)
    Dense(1, activation='sigmoid')             # Output layer (0 or 1)
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=1000, verbose=0)  # Train for 1000 epochs

# Test the model
predictions = model.predict(X)
print("XOR Predictions:")
for i in range(len(X)):
    print(f"{X[i]} → {np.round(predictions[i][0])} (expected: {y[i][0]})")
```
